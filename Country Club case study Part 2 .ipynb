```python
import sqlite3
from sqlite3 import Error

 
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
 
    return conn
```

### Q10: Produce a list of facilities with a total revenue less than 1000. The output of facility name and total revenue, sorted by revenue. Remember that there's a different cost for guests and members!

```python
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
        SELECT Facilities.name, SUM(Bookings.slots * (
        CASE WHEN Bookings.memid = 0 THEN Facilities.guestcost ELSE Facilities.membercost END)) AS totalrevenue
        FROM Bookings
        INNER JOIN Facilities 
        ON Bookings.facid = Facilities.facid
        GROUP BY Facilities.name
        HAVING SUM(Bookings.slots * (
        CASE WHEN Bookings.memid = 0 THEN Facilities.guestcost ELSE Facilities.membercost END)) < 1000
        ORDER BY totalrevenue; 
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("2. Query all tasks")
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```

###  Q11: Produce a report of members and who recommended them in alphabetic surname,firstname order

```python
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
        SELECT m.firstname || ' ' || m.surname AS FullName, (
            SELECT m.firstname || ' ' || m.surname
            FROM Members AS m
            WHERE recommendedby = m.memid) AS recommended
        FROM Members AS m
        ORDER BY FullName;
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("2. Query all tasks")
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```

### Q12: Find the facilities with their usage by member, but not guests

```python
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
        SELECT Facilities.name, COUNT(Facilities.name) AS usage
        FROM Bookings 
        INNER JOIN Facilities 
        ON Bookings.facid = Facilities.facid
        INNER JOIN Members
        ON Bookings.memid = Members.memid
        WHERE Bookings.memid != 0.0
        GROUP BY name;
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("2. Query all tasks")
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```

### Q13: Find the facilities usage by month, but not guests

```python
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
        SELECT strftime('%m', Bookings.starttime) AS month, Facilities.name, COUNT(Facilities.name)
        FROM Bookings 
        LEFT JOIN Facilities 
        ON Bookings.facid = Facilities.facid
        WHERE Bookings.memid != 0.0
        GROUP BY month, name;
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("2. Query all tasks")
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```
